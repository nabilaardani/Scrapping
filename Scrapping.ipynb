```python
url = "http://telpon.info/bengkel-motor/surabaya/"
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.text, "html.parser")
print(req.text)
```

```python
import pandas as pd

df = pd.DataFrame()

for div in soup.find_all('div', class_='company-listing clearfix'):
    nama = div.find_all('div', class_='listing-title')
    alamat = div.find_all('div', class_='detail address')
    telp = div.find_all('div', class_='detail phone')

    df = df.append({'nama bengkel': nama, 'alamat': alamat, 'no telp': telp}, ignore_index=True)

    print(div, '\n\n')
```

```python
url = "https://www.autofun.co.id/bengkel-motor/detail-honda-di-surabaya-313288"
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.text, "html.parser")
print(req.text)
```

```python
import pandas as pd

df = pd.DataFrame()

for div in soup.find_all('a', class_='review-item wa-g-unit'):
    nama = div.find_all('span', class_='wa-ellipsis-1')
    ulasan = div.find_all('p')[-1].getText()

    df = df.append({'Nama': nama, 'Ulasan': ulasan}, ignore_index=True)

    print(div,'\n\n')
```

```python
url = "https://iws-motor-sport.business.site/"
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.text, "html.parser")
print(req.text)
```

```python
import pandas as pd

df = pd.DataFrame()

for div in soup.find_all('div', class_='EIjale'):
    nama = div.find_all('cite')[0].getText()
    ulasan = div.find_all('blockquote')[-1].getText()

    df = df.append({'Nama': nama, 'Testimoni': ulasan}, ignore_index=True)

    print(div,'\n\n')
```
